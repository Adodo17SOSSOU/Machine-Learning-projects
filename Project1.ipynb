As the result is showing us, this dataset is highly unbanlanced and if we use it like that, our model will not perform well. 0 means Normal transaction and 1 means fraudulent transaction

```python
#Now, we're going to separate our data for analysis. This means separate the fraudulent part from the normal one
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
#Let's see the shape of both the fraudulent and the normal data 
print("The shape of the normal data is :")
print(legit.shape)
print("The shape of the fraudulent data is :")
print(fraud.shape)
```

Under-Sampling of the data

We are going to normalyze the dataset. This means we are going to build a sample dataset containing similar distribution of normal transaction and the fraudulent one.

The number of Fraudulent transaction is 492 

#Let's take randomly the same number of rows from the normal transaction so that our dataset will contain exactly 492 of normal data and 492 of fraudulent data


```python
legit_sample=legit.sample(n=492)
```

Now let's us concatenate the 2 datasets to form the final dataset we are going to use

```python
new_dataset=pd.concat([legit_sample,fraud], axis=0)
```

```python
#Now let us see the first rows of that new dataset
new_dataset.head()
```

```python
#The last rows
new_dataset.tail()
```

```python
#Now we are going to verify if we have the same number of normal transaction 492 which is the number of fraudulent transaction
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Now we can split our dataset into Features and Targets

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']

#Print X 
print(X)
```

```python
#print Y
print(Y)
```

Now let's split our data into training sample and testing sample

```python
X_train,X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
#The shape of our training, testing and our actual dataset
print(X.shape, X_train.shape, X_test.shape)
```

Now it's time for Model training

Logistic Regression because it is about deciding if or not a transaction is fraudulent. 1 or 0

```python


# Construct a pipeline with preprocessing and logistic regression
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('logistic', LogisticRegression())
])

# Define parameter grid for GridSearchCV
param_grid = {
    'logistic__max_iter': [500, 1000, 1500],
    'logistic__solver': ['lbfgs', 'sag', 'saga']
}


# Perform grid search with cross-validation
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, Y_train)

# Print best parameters and score
print("Best parameters:", grid_search.best_params_)
print("Best cross-validation accuracy:", grid_search.best_score_)
best_model = grid_search.best_estimator_
```

```python
#But what did the model predicted for the training data?

test_accuracy = best_model.score(X_train, Y_train)
test_accuracy_training = best_model.score(X_train, Y_train)
print("training set accuracy :", test_accuracy)
```

```python
# Evaluate the best model on the test set
test_accuracy = best_model.score(X_test, Y_test)
print("Test set accuracy:", test_accuracy)
```

That's great, the accuracy score for the training and the test sample are awesome and are very similar, So, the model is performing very well
