```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#Let's load our dataset 
diabets_dataset=pd.read_csv('diabetes.csv')
```

```python
#The first lines in our dataset
diabets_dataset.head()
```

```python
#The last lines in our dataset
diabets_dataset.tail()
```

```python
#The shape of our dataset--->Number of rows and columns in the dataset
print(diabets_dataset.shape)
```

```python
#The statistical measures of the data
diabets_dataset.describe()
```

```python
#How many diabet people do we have in out dataset???
diabets_dataset['Outcome'].value_counts()
```

```python
#Now we are separating the dataset into the features and the labels
X=diabets_dataset.drop(columns='Outcome',axis=1)
Y=diabets_dataset['Outcome']
```

```python
#Print X 
print(X)
```

```python
#print Y
print(Y)
```

Data Standardization

```python
scaler=StandardScaler()
```

```python
standardized_data=scaler.fit_transform(X)
```

```python
print(standardized_data)
```

```python
X=standardized_data
Y=diabets_dataset['Outcome']  #Already done anyway
```

```python
print(X)
```

```python
print(Y)
```

Now we need to split our data into training data and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Now we can train our model

```python
classifier=svm.SVC(kernel='linear')
```

```python
#We're gonna train then the support vector Machine Classifier
classifier.fit(X_train,Y_train)
```

Now let's evaluate our Model

```python
#How does the model perform on the training data?
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('The accuracy score of our training data is : ',training_data_accuracy)
```

```python
#What about our test data?
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('The accuracy score of our test data is : ',test_data_accuracy)
```

We actually have a good accuracy score !!!

Now we can make a predictive system

```python
import warnings
warnings.filterwarnings("ignore", message="X does not have valid feature names")

input_data=(3,158,76,36,245,31.6,0.851,28) # This is just an example

#let us change it to a numpy array
input_array=np.asarray(input_data)

#We should reshape the array as we are predicting for only one instance

input_reshaped=input_array.reshape(1,-1)

#Standardize the input data

std_data=scaler.transform(input_reshaped)
print(std_data)

prediction=classifier.predict(std_data)

print('The prediction is :',prediction)

#Just print a message saying if the input data corresponds to a diabetic data or a non diabetic person's data

if (prediction[0]==0):
    print('The person is not a diabetic')
else:
    print("The person is a diabetic")
```
