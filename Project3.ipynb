```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Let's load our dataset 
heart_data=pd.read_csv('heart.csv')
```

```python
#Print the first rows of our dataset
heart_data.head()
```

```python
#The last rows of our dataset
heart_data.tail()
```

```python
#The shape of our dataset->the number of rows and columns
heart_data.shape
```

```python
#Some informations about our dataset
heart_data.info()
```

```python
#Some descriptions about the dataset
heart_data.describe()
```

```python
#Checkinf if there are some missing values in the dataset
heart_data.isnull().sum()
```

```python
#How many people have heart disease and how many do not have it?
heart_data['target'].value_counts()
```

In out dataset, 526 persons have a heart desease and 499 do not

Splitting the features and Target

```python
X=heart_data.drop(columns='target', axis=1)
Y=heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into training data and test data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Now we are going to train our Model

```python
model = LogisticRegression(max_iter=1000)
```

```python
#Training the Logistic Regression model with our training data
model.fit(X_train,Y_train)
```

Model Evaluation

```python
# Now what is the accuracy score on the training data?? 
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy score on the training data is:',training_data_accuracy)
```

```python
#Now let's see how the model is performing on our test data
# Now what is the accuracy score on the training data?? 
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy score on the test data is:',test_data_accuracy)
```

Our model is performing greatly !!!

Now we can build a predictive system 

```python
import warnings
warnings.filterwarnings("ignore", message="X does not have valid feature names")

input_data=(51,1,0,140,298,0,1,122,1,4.2,1,3,3) # We can change this with any other input

#let us change it to a numpy array
input_array=np.asarray(input_data)

#We should reshape the array as we are predicting for only one instance

input_reshaped=input_array.reshape(1,-1)

prediction=model.predict(input_reshaped)

print('The prediction is :',prediction)

#Just print a message saying if the input data corresponds to a data of someone with heart desease or not

if (prediction[0]==0):
    print('The person does not have a heart desease!')
else:
    print("The person has a heart desease !")
```
